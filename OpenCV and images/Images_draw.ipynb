# Drawing on Images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
blank_img = np.zeros((512,512,3),dtype=np.int16)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img);
```

# Shapes

### Rectangles

* img Image.
* pt1 Vertex of the rectangle.
* pt2 Vertex of the rectangle opposite to pt1 .
* color Rectangle color or brightness (grayscale image).
* thickness Thickness of lines that make up the rectangle. Negative values, like #FILLED,mean that the function has to draw a filled rectangle.
* lineType Type of the line. See #LineTypes
* shift Number of fractional bits in the point coordinates.

```python
cv2.rectangle(blank_img, pt1=(380,10),pt2=(500,128),color=(0,255,0),thickness=10) # green rectangle
plt.imshow(blank_img);
```

```python
cv2.rectangle(blank_img, pt1=(10,500),pt2=(100,400),color=(255,0,0),thickness=10) # red rectangle
plt.imshow(blank_img);
```

```python
cv2.rectangle(blank_img, pt1=(200,300), pt2=(300,200), color=(0,0,255), thickness=10) # blue square
plt.imshow(blank_img);
```

```python
cv2.circle(blank_img, center=(100,100), radius=50, color=(255,0,255), thickness=8) # pink cercle
plt.imshow(blank_img);
```

```python
cv2.circle(blank_img, center=(400,400), radius=50, color=(255,255,0), thickness=-1) # fill cercle
plt.imshow(blank_img);
```

```python
cv2.line(blank_img, pt1=(0,0), pt2=(512,512), color=(102,255,255), thickness=5)
plt.imshow(blank_img);
```

### Text

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img, text='OMG', org=(10,300), fontFace=font, fontScale=4,
            color=(255,255,255), thickness=3, lineType=cv2.LINE_AA )
plt.imshow(blank_img);
```

### Polygons

To draw a polygon, first you need coordinates of vertices. Make those points into an array of shape ROWSx1x2 where ROWS are number of vertices and it should be of type int32. 

```python
vertices = np.array([[100,300] ,[200,200], [400,300], [200,400]], dtype=np.int32)
```

```python
vertices.shape
```

```python
pts = vertices.reshape((-1,1,2))
```

```python
pts.shape
```

```python
vertices
```

```python
pts
```

```python
cv2.polylines(blank_img, [pts], isClosed=True, color=(255,0,0), thickness=5)
plt.imshow(blank_img);
```
